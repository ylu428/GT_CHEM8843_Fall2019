```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB ##use .sigma_ to get variance for each dimension for each class.
from sklearn.datasets import make_blobs
```

```python
from sklearn import datasets
```

```python
X,y=datasets.make_blobs(1000,2,centers=3)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=y)

rng=np.random.RandomState(0)
model = GaussianNB()
model.fit(X,y)
Xnew = [-5,-5]+[18,18]*rng.rand(2000,2)
ynew = model.predict(Xnew)
plt.scatter(Xnew[:,0],Xnew[:,1],c=ynew,s=20,alpha = 0.1)
plt.show()
```

```python
plt.scatter(Xnew[:,0],Xnew[:,1],c=model.predict_proba(Xnew)[:,0],s=20)
plt.title("Color indicates proability of being in a given class")
plt.show()
plt.scatter(Xnew[:,0],Xnew[:,1],c=model.predict_proba(Xnew)[:,1],s=20)
plt.title("Color indicates proability of being in a given class")
plt.show()
plt.scatter(Xnew[:,0],Xnew[:,1],c=model.predict_proba(Xnew)[:,2],s=20)
plt.title("Color indicates proability of being in a given class")
plt.show()
```
