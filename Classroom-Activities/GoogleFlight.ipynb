```python
import requests as rq
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from pandas import ExcelWriter
from dateutil import parser
import datetime
import time


airport =[]     ### create dictionary to hold ticker symbols
wikiURL = 'https://zh.wikipedia.org/wiki/%E5%85%A8%E7%90%83%E6%9C%BA%E5%9C%BA%E5%AE%A2%E8%BF%90%E5%90%9E%E5%90%90%E9%87%8F%E5%88%97%E8%A1%A8'
wikiResponse = rq.get(wikiURL)      ##use requests library to retreive info from webpage
tst = wikiResponse.text       #just get the text from the requests object
text = tst.split("<tr>")
```

```python
for i in range(2,41):
    tick=text[i].split('<td>')
    ticker = tick[4].split('</td>') # the forth item in tick is the symbol of airports
    airport.append(ticker[0])## append each ticker symbol to our dictionary
```

```python
flight_info={'departure airport':[], 
             'destination':[], 
             'departure date':[], 
             'returning date':[],
             'lowest price':[],
             'google flight':[]}

driver=webdriver.Chrome()
```

```python
dDate = ['2020-05-27','2019-11-24']
rDate = ['2020-06-03','2019-11-30']

deltaT = 1
pluseMinusDays = 2*deltaT+1

# make a list for depature date base on dDate
dDay = ['']*pluseMinusDays*len(dDate)
d = 0
for ddate in dDate:
    for timeD in range(pluseMinusDays):
        dDay[d+timeD]=(parser.parse(ddate)+datetime.timedelta(days=(timeD-deltaT))).strftime('%Y-%m-%d')
    d +=pluseMinusDays    

# make a list for return date base on rDate
rDay = ['']*pluseMinusDays*len(rDate)
r = 0
for rdate in rDate:
    for timeR in range(pluseMinusDays):
        rDay[r+timeR]=(parser.parse(rdate)+datetime.timedelta(days=(timeR-deltaT))).strftime('%Y-%m-%d')
    r +=pluseMinusDays

for rdate in rDay:
    for ddate in dDay:
        if ddate < rdate:
            for j in range(len(airport[:2])):
                for i in range(len(airport[:2])):
                    if airport[i] != airport[j]:
                        url = 'https://www.google.com/flights?hl=en#flt='+airport[j]+'.'+airport[i]+'.'+ddate+'*'+airport[i]+'.'+airport[j]+'.'+rdate+';c:USD;e:1;sd:1;t:f'
                        flight_info['departure airport'].append(airport[i])
                        flight_info['destination'].append(airport[j])
                        flight_info['departure date'].append(ddate)
                        flight_info['returning date'].append(rdate)
                        flight_info['google flight'].append(url)
                        driver.get(url)
                        time.sleep(6)
                        price =driver.find_element_by_xpath('//*[@id="flt-app"]/div[2]/main[4]/div[7]/div[1]/div[5]/div[1]/ol/li[1]/div/div[1]/div[2]/div[1]/div[1]/div[5]/div[1]')
                        flight_info['lowest price'].append(price.text)

flight_info_df = pd.DataFrame(flight_info)
flight_info_df
```
